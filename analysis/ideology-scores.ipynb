```python
import json
from pandas.io.json import json_normalize
import pandas as pd
import altair as alt
```

```python
with open('lawmakers-2019.json') as f:
    raw = json.load(f)
df = json_normalize(raw, 'lawmakers')
```

```python
df['district_acres'] = df['district_acres'].astype(str).str.replace(',','').astype(int)
house = df[df['chamber'] == 'house']
senate = df[df['chamber'] == 'senate']
```

```python
alt.Chart(house).mark_point().encode(
#     alt.Y('name:N', sort=alt.EncodingSortField(field='ideologyScore')),
    alt.Y('district_acres', scale=alt.Scale(type='log')),
    alt.X('ideologyScore:Q'),
    alt.Color('party'),
    alt.Tooltip(['name'])
).configure(background='#fff')
```
